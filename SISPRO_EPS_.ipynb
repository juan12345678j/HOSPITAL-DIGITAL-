```python
import pandas as pd

# Función para cargar el archivo y verificar columnas
def cargar_y_verificar(file_path, required_columns):
    df = pd.read_excel(file_path)
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"Las siguientes columnas no se encuentran en el DataFrame: {', '.join(missing_columns)}")
    return df

# Función para calcular la edad en días entre dos fechas
def calcular_edad_en_dias(df, fecha_nacimiento_col, fecha_atencion_col):
    # Asegúrate de que las columnas de fechas estén en formato datetime
    df[fecha_nacimiento_col] = pd.to_datetime(df[fecha_nacimiento_col], errors='coerce')
    df[fecha_atencion_col] = pd.to_datetime(df[fecha_atencion_col], errors='coerce')

    # Calcula la diferencia en días
    df['EdadEnDias'] = (df[fecha_atencion_col] - df[fecha_nacimiento_col]).dt.days

    return df

# Función para calcular conteos y porcentajes
def calcular_estadisticas(df, edad_min, edad_max, columnas_datos):
    filtro = (df['EdadEnMeses'].between(edad_min, edad_max)) & (df['UltimoSeguimiento'] == True)
    conteo = df[filtro].shape[0]

    resultados = {}
    for columna in columnas_datos:
        if columna in df.columns:
            conteo_filas = df[filtro & df[columna].notna()].shape[0]
            porcentaje = (conteo_filas / conteo) * 100 if conteo > 0 else 0
            resultados[columna] = {'conteo': conteo_filas, 'porcentaje': porcentaje}
        else:
            resultados[columna] = {'conteo': 0, 'porcentaje': 0}
            print(f"La columna '{columna}' no se encuentra en el DataFrame")

    return conteo, resultados

# Función para imprimir resultados
def imprimir_resultados(edad_min, edad_max, conteo, resultados):
    print(f"Número de filas con edad entre {edad_min} y {edad_max} meses y UltimoSeguimiento = True: {conteo}")
    for columna, stats in resultados.items():
        print(f"Número de filas con datos en {columna}: {stats['conteo']}")
        print(f"Porcentaje de filas con datos en {columna} sobre el total: {stats['porcentaje']:.2f}%")

# Función para imprimir estadísticas adicionales
def imprimir_estadisticas_adicionales(df):
    # Filtros para análisis adicionales
    filtro_edad_mes = (df['EdadEnMeses'].notna()) & (df['UltimoSeguimiento'] == True)
    conteo_edad_mes = df[filtro_edad_mes].shape[0]

    if 'FechaUltimoControl' in df.columns:
        conteo_fecha_ultimo_control = df[filtro_edad_mes & df['FechaUltimoControl'].notna()].shape[0]
        porcentaje_fecha_ultimo_control = (conteo_fecha_ultimo_control / conteo_edad_mes) * 100 if conteo_edad_mes > 0 else 0
        print(f"Número de filas con FechaUltimoControl no nula: {conteo_fecha_ultimo_control}")
        print(f"Porcentaje de filas con FechaUltimoControl no nula sobre las filas con EdadEnMeses no nula: {porcentaje_fecha_ultimo_control:.2f}%")

    filtro_edad_dias = (df['EdadEnDias'].between(0, 30)) & (df['UltimoSeguimiento'] == True)
    conteo_edad_dias = df[filtro_edad_dias].shape[0]

    if 'FechaAtencionLactancia' in df.columns:
        conteo_fecha_atencion_lactancia = df[filtro_edad_dias & df['FechaAtencionLactancia'].notna()].shape[0]
        porcentaje_fecha_atencion_lactancia = (conteo_fecha_atencion_lactancia / conteo_edad_dias) * 100 if conteo_edad_dias > 0 else 0
        print(f"Número de niños con EdadEnDias entre 0 y 30 días y FechaAtencionLactancia no nula: {conteo_fecha_atencion_lactancia}")
        print(f"Porcentaje de niños con EdadEnDias entre 0 y 30 días y FechaAtencionLactancia no nula: {porcentaje_fecha_atencion_lactancia:.2f}%")

    # Contar el número de niños con EdadEnDias entre 0 y 30 días y UltimoSeguimiento = True
    print(f"Número de niños con EdadEnDias entre 0 y 30 días y UltimoSeguimiento = True: {conteo_edad_dias}")

# Parámetros del archivo
file_path = '/content/SeguimientoDNT__1_con_edad_en_meses.xlsx'

# Cargar el DataFrame y verificar columnas
df = cargar_y_verificar(file_path, ['EdadEnMeses', 'UltimoSeguimiento', 'FechaNacimiento', 'FechaAtencion'])

# Calcular la edad en días y agregarla al DataFrame
df = calcular_edad_en_dias(df, 'FechaNacimiento', 'FechaAtencion')

# Definición de análisis
analisis = [
    {'edad_min': 6, 'edad_max': 23, 'columnas': ['FechaTamizaje', 'AnemiaFerropenica']},
    {'edad_min': 6, 'edad_max': 24, 'columnas': ['FechaMicronutientes']},
    {'edad_min': 24, 'edad_max': 59, 'columnas': ['FechaSuplementacionVitaminaA']},
    {'edad_min': 24, 'edad_max': 59, 'columnas': ['FechaSuplementacionHierro']},
    {'edad_min': 12, 'edad_max': 23, 'columnas': ['FechaDesparasitacion']},
    {'edad_min': 24, 'edad_max': 48, 'columnas': ['FechaDesparasitacion']}
]

# Ejecutar análisis
for analisis_item in analisis:
    conteo, resultados = calcular_estadisticas(df, analisis_item['edad_min'], analisis_item['edad_max'], analisis_item['columnas'])
    imprimir_resultados(analisis_item['edad_min'], analisis_item['edad_max'], conteo, resultados)

# Imprimir estadísticas adicionales
imprimir_estadisticas_adicionales(df)
```
